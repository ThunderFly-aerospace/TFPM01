```python
from pyulog import ULog
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
u = ULog('/mnt/data/ownCloudMlab/TF_data/UAV/TF-G2/zkousky/Test_20201124_zdiby/log/log_24_2020-11-24-14-18-44.ulg')
```

```python
t = None
for x in u.data_list:
    print(x.name)
    if x.name == 'tunnel':
        t = x
        print(".....")
```

```python
if t:
    df = pd.DataFrame.from_dict(t.data)
```

```python
df['nc_0.5'] = df.apply(lambda row: row['payload[0]'] + (int(row['payload[1]']) << 8) 
                        + (int(row['payload[2]']) << 16) + (int(row['payload[3]']) << 24), axis=1)
df['nc_1'] = df.apply(lambda row: row['payload[4]'] + (int(row['payload[5]']) << 8) 
                        + (int(row['payload[6]']) << 16) + (int(row['payload[7]']) << 24), axis=1)
df['nc_2.5'] = df.apply(lambda row: row['payload[8]'] + (int(row['payload[9]']) << 8) 
                        + (int(row['payload[10]']) << 16) + (int(row['payload[11]']) << 24), axis=1)
df['nc_4'] = df.apply(lambda row: row['payload[12]'] + (int(row['payload[13]']) << 8) 
                        + (int(row['payload[14]']) << 16) + (int(row['payload[15]']) << 24), axis=1)
df['nc_10'] = df.apply(lambda row: row['payload[16]'] + (int(row['payload[17]']) << 8) 
                        + (int(row['payload[18]']) << 16) + (int(row['payload[19]']) << 24), axis=1)

df['mc_1'] = df.apply(lambda row: row['payload[20]'] + (int(row['payload[21]']) << 8) 
                        + (int(row['payload[22]']) << 16) + (int(row['payload[23]']) << 24), axis=1)
df['mc_2.5'] = df.apply(lambda row: row['payload[24]'] + (int(row['payload[25]']) << 8) 
                        + (int(row['payload[26]']) << 16) + (int(row['payload[27]']) << 24), axis=1)
df['mc_4'] = df.apply(lambda row: row['payload[28]'] + (int(row['payload[29]']) << 8) 
                        + (int(row['payload[30]']) << 16) + (int(row['payload[31]']) << 24), axis=1)
df['mc_10'] = df.apply(lambda row: row['payload[32]'] + (int(row['payload[33]']) << 8) 
                        + (int(row['payload[34]']) << 16) + (int(row['payload[35]']) << 24), axis=1)

df['tps'] = df.apply(lambda row: row['payload[36]'] + (int(row['payload[37]']) << 8) 
                        + (int(row['payload[38]']) << 16) + (int(row['payload[39]']) << 24), axis=1)
```
